```python
def covid():
    dfcovid = pd.read_csv("COVID_dataset_V0.csv")
    sorteddf = dfcovid.sort_values(by = [])
    return dfcovid
covid()
```

```python
def college_id():
    dfcollege = pd.read_csv("Most-Recent-Cohorts-Institution.csv", low_memory = False)
    dfcollege = dfcollege.iloc[:,0:4]
    dfsimple = dfcollege.drop(columns = ["OPEID","OPEID6"])
    dfsimplecollege = dfsimple.rename(columns = {"UNITID": "unitid"})
    return dfsimplecollege
college_id()
```

```python
def combine():
    dfcovid = covid()
    dfsimplecollege = college_id()
 
    done = pd.merge(dfcovid, dfsimplecollege, on = "unitid", how = "outer")
    done = done[["INSTNM", "county_name", "city","state", "zip", "latitude","longitude","date","week","admission_rate","number_students","TotalCount","TotalCountWeekly","TotalConfirmedPositive","TotalConfirmedPositiveWeekly","TotalPositivityRate","TotalPositivityRateWeekly"]]
    done = done.iloc[:1130]
    real_done = done.rename(columns={"INSTNM":"college", "TotalCount":"total_count", "TotalCountWeekly":"total_count_weekly","TotalConfirmedPositive":"total_confirmed_positive","TotalConfirmedPositiveWeekly":"total_confirmed_positive_weekly","TotalPositiveRate":"total_positive_rate","TotalPositivityRateWeekly":"total_positivity_rate_weekly"})
#     with pd.ExcelWriter("blankSheet.xlsx") as writer:
#         writer.book = openpyxl.load_workbook("blankSheet.xlsx")
#         real_done.to_excel(writer, sheet_name='college')
    with pd.ExcelWriter('college.xlsx',mode='w') as writer:  
        real_done.to_excel(writer)
#     for i in range(len(real_done)):
#         collegename = real_done[i,"college"]
#         print(college)
    return real_done

combine()
```

```python

def hash():
    df = pd.read_excel('college.xlsx')
    college = {}
    for index, row in df.iterrows():
        college.update({row["college"]: row["state"]})

    return college
hash()
```
